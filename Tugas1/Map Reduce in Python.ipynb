```python
#Import the required libraries
import sys
from itertools import groupby
```

```python
# Create a map function
# In this function, we will count every word as 1
def mapfunc(w):
    cleanword = ''.join([i for i in w if i.isalpha()])
    return [cleanword, 1]
```

```python
# Create a reduce function
# We're going to map every same words and increase the total words
def reducefunc(key, values):
    counts = [x[1] for x in values]
    return [key,sum(counts)]
```

```python
#Split the document into single word
with open("greatgatsby.txt") as f:
    words=[word for line in f for word in line.split()]
```

```python
print(words)
```

```python
# Using the map function in python to help mapping each words with the format (word, 1)
map_result = map(mapfunc, words)
```

```python
# Sort the map result lexigographically by the word
map_result_sorted = sorted (map_result, key = lambda x: x[0])
```

```python
reduce_result = []
```

```python
# Making a list of words and their total count in the source text file
for k,g in groupby(map_result_sorted,key = lambda x:x[0]):
    reduce_result.append(reducefunc(k, list(g)))
reduce_result
```
